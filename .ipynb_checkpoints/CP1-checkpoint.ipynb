```python
# References:
# https://github.com/aimacode/aima-python/blob/master/search.ipynb

# Node Class
class Node():
    def __init__(self, state, edgeWeight=0):
        self.state = state # Letter
#         self.parent = parent
        self.edgeWeight = edgeWeight # Cost from parent to self
        self.children = []
        '''
        # Need depth?
        self.depth = 0
        if parent:
            self.depth = parent.depth + 1
        '''
    def returnChildren(self):
        return self.children
    
    def returnState(self):
        return self.state
    
    def returnEdgeWeight(self):
        return self.edgeWeight
    
    # Assigns the child's parent & edgeweight + appends childNode to self.children
    def addChild(self, childNode, edgeWeight):
        childNode.parent = self
        childNode.edgeWeight = edgeWeight
        self.children.append(childNode)
        
    # Returns the list of states from root to current + the total edge weight
    def getPath(self):
        node = self
        pathBack = []
        totalEdgeWeight = 0
        while node:
            pathBack.append(node.state)
            totalEdgeWeight += node.edgeWeight
            node = node.parent
        return [list(reversed(pathBack)), totalEdgeWeight]

# Tree Class
class Tree():
    def __init__(self):
        self.numberOfNodes = 0
        self.root = None
        self.path = []
        self.visited = []
        self.destinationFound = False

    def addRoot(self, rootNode):
        self.root = rootNode
        self.numberOfNodes +=1

    def addNode(self, childNode, parentNode, edgeWeight):
        #if no nodes, add root node
        if self.numberOfNodes == 0:
            self.root = parentNode
        else : 
            parentNode.addChild(childNode, edgeWeight)
        self.numberOfNodes +=1

            
        
    
    # Depth-first search, Breadth-first search, or Uniform cost search
    # Searching for the first instance of (node.state == "G")
    def search(self, currNode, destination):

        if(self.destinationFound):
            return
        
        self.visited.append(currNode.state)
        self.path.append(currNode.state)
        
        if(currNode.state == destination.state):
            self.destinationFound = True
            return
        
        #adding children to frontier
        for n in currNode.returnChildren():
            self.search(n, destination)
            if(not self.destinationFound):
                self.path.pop()
        
        
# In main
tree = Tree()
root = Node("S")
tree.addRoot(root)
a = Node("A")
b = Node("B")
c = Node("C")
d = Node("D")
e = Node("E")
g = Node("G")

tree.addNode(a, tree.root, 3)
tree.addNode(b, tree.root, 1)
tree.addNode(c, tree.root, 8)
tree.addNode(d, a, 3)
tree.addNode(e, a, 7)
tree.addNode(g, a, 15)
tree.addNode(g, b, 20)
tree.addNode(g, c, 5)

tree.search(tree.root, g)

print(tree.path)
print(tree.visited)


    
```
